## Вводные по задаче.



Банки используют модели кредитного риск-менеджмента, чтобы понимать,
насколько можно доверять клиенту в выполнении обязательств по договорам
кредитования. Когда вы как клиент заполняете заявку на кредит или ипотеку,
вас оценивают по модели кредитного риск-менеджмента. Банк может
использовать разные сведения: например, о месте работы, возрасте, истории
предыдущих погашений по другим кредитам в банках и кредитных
организациях. На основе этой информации модель машинного обучения
подсказывает кредитному менеджеру, стоит ли вам доверять запрашиваемую
сумму денег.
С помощью такой автоматизации банк экономит время своих специалистов,
чтобы они не искали и не агрегировали информацию по каждому клиенту для
принятия решения о выдаче кредита. Это ускоряет время подтверждения
заявки на кредит. Однако в отдельных случаях специалисты могут экспертно
проверить решение модели, чтобы проаудировать её и выявить возможные
слабые места.

#### Проблема, которую предстоит решить:

В рамках проекта решить востребованную задачу — оценить
риск неуплаты клиента по кредиту (дефолт).

Дефолт — неуплата процентов по кредиту или облигациям, непогашение
займа в течение определённого времени t. Обычно дефолт считают
свершившимся, если клиент не совершил выплату по кредиту в течение
90 дней.

Нужная модель позволяет банку или другой кредитной организации оценить
текущий риск по любым выданным займам и кредитным продуктам
и с большей долей вероятности предотвратить неисполнение кредитных
обязательств клиентом. Таким образом, банк меньше рискует понести
убытки.

```python




def read_parquet_dataset_from_local(path_to_dataset: str, start_from: int = 0,
                                     num_parts_to_read: int = 2, columns=None, verbose=False) -> pd.DataFrame:
    """
    Reads num_parts_to_read partitions, converts them to pd.DataFrame and returns.
    """
    dataset_paths = sorted([os.path.join(path_to_dataset, filename) for filename in os.listdir(path_to_dataset)
                             if filename.startswith('train')])

    start_from = max(0, start_from)
    chunks = dataset_paths[start_from: start_from + num_parts_to_read]

    if not chunks:
        return pd.DataFrame()  # Return empty DataFrame if no chunks.

    if verbose:
        print('Reading chunks:\n')
        for chunk in chunks:
            print(chunk)

    dataframes = []
    for chunk_path in tqdm.tqdm(chunks, desc="Reading dataset with pandas"):
        df = pd.read_parquet(chunk_path, columns=columns)
        dataframes.append(df)

    # Concatenate only if there are dataframes to concatenate
    return pd.concat(dataframes, ignore_index=True) if dataframes else pd.DataFrame()

def prepare_transactions_dataset(path_to_dataset: str, num_parts_to_preprocess_at_once: int = 1,
                                 num_parts_total: int = 50, save_to_path=None, verbose: bool = False):
    """
    Prepares and aggregates transactions data into a ready DataFrame.
    """
    # значения целевой переменной
    targets = pd.read_csv(target_path)
    # Переименуем целевую переменную - target
    targets.rename(columns={'flag': 'target'}, inplace=True)

    final_df = pd.DataFrame()  # For final aggregation

    for step in tqdm.tqdm(range(0, num_parts_total, num_parts_to_preprocess_at_once),
                          desc="Transforming transactions data"):
        transactions_frame = read_parquet_dataset_from_local(path_to_dataset, start_from=step,
                                                             num_parts_to_read=num_parts_to_preprocess_at_once,
                                                             verbose=verbose)

        if transactions_frame.empty:
            continue  # Skip to the next step if no data

        # Агрегация по mean, int
        # Create aggregation dictionary
        agg_d = {col: 'mean' for col in transactions_frame.columns if col not in ['id', 'rn']}
        agg_d['rn'] = 'count'  # Count 'rn'

        # Grouping and aggregating data
        prepared_df = transactions_frame.groupby('id').agg(agg_d).reset_index(drop=False).astype(int)


        # Фильтрация по rn - max
        #df_filtered = transactions_frame.groupby('id')['rn'].min().reset_index()
        #df_filtered = pd.merge(df_filtered, transactions_frame, on=['id', 'rn'], how='left')
        #df_filtered = prepared_df.drop_duplicates(subset=['id'], keep='first')

        # Append to final DataFrame
        final_df = pd.concat([final_df, prepared_df], ignore_index=True)

    # Объединяем с целевыми значениями
    final_df = final_df.merge(targets, on='id', how='inner')


    #if save_to_path:
    #    final_df.to_parquet(os.path.join(save_to_path, 'research_dfparquet'), index=False)

    return final_df  # Return the final aggregated DataFrame

# Example usage
#data = prepare_transactions_dataset(path, num_parts_to_preprocess_at_once=2, num_parts_total=50,
 #                                     save_to_path='data/traindata/')
#                                      save_to_path = '/content/drive/MyDrive/Skillbox/Advanced/credit risk_management/data')
```

#### Data cleaning

```python



# Дубликаты столбцов
def column_duplicates(df):
    duplicate_columns = []
    columns = df.columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            if df[columns[i]].equals(df[columns[j]]):
                duplicate_columns.append((columns[i], columns[j]))

    print("Пары одинаковых столбцов:")
    for col1, col2 in duplicate_columns:
        print(f"{col1} и {col2} имеют одинаковые значения.")




# Удаление дубликатов target==0
def duplicates(df):
    # Удаляем id
    df.drop(['id'], axis=1, inplace=True, errors='ignore')# Удаление дубликатов target==0
    # контроль целевой переменной
    print('до', df['target'].value_counts())

    # создаем маску для дубликатов, где target == 1
    mask = df.duplicated(keep=False) & (df['target'] == 1)

    # Оставляем дубликаты, где target == 1
    df_keep_target_1 = df[mask]

    # Удаляем все дубликаты, кроме тех, которые соответствуют маске
    df_unique = df[~mask].drop_duplicates()

    # Объединяем оба DataFrame
    df = pd.concat([df_unique, df_keep_target_1]).sort_index()

    print('после', df['target'].value_counts())
    print('Размер DataFrame:', df.shape)

    return df




# Количество пропусков
def missings(df):
    missing_values = df.isna().sum()
    # Получаем процент пропусков
    missing_percentage = (missing_values / df.shape[0]) * 100
    # Объединяем в один DataFrame
    missing_info = pd.DataFrame({
        'Количество пропусков': missing_values,
        'Процент пропусков': missing_percentage
        })
    # Создаем переменную с пропусками
    missing_info = missing_info[missing_info['Количество пропусков'] > 0]

    return missing_info


# Расчет квантилей
def calculate_iqr_boundaries(series):
  q25 = series.quantile(0.25)
  q75 = series.quantile(0.75)
  iqr = q75 - q25

  boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
  return boundaries
```

#### Важность признаков и корреляции

```python

# Оценка корреляций признаков к target
def col_min_corr(df):
    min_corr = []
    for i in df.columns:
        j = df[i]
        corr = df['target'].corr(j)
        if -0.01 < corr and corr < 0.01:
            min_corr.append(i)

    return min_corr




# Оценка значимости признаков
def importantes_0(df):
    # Определение данных
    X = df.drop(['target'], axis=1)
    y = df['target']

    # Обучение модели
    params = {
        'objective': 'binary:logistic',
        'max_depth': 3,
        'learning_rate': 0.1
        }
    model = xgboost.train(params, xgboost.DMatrix(X, y), num_boost_round=10)

    # Создание объекта Explainer и получение SHAP значений
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X)

    # Преобразуем shap_values в DataFrame для удобной работы
    shap_df = pd.DataFrame(shap_values, columns=X.columns)

    # Создание переменной no_good для признаков с нулевой значимостью
    no_good = shap_df.mean().index[shap_df.mean() == 0].tolist()

    # Выводим список признаков с нулевой значимостью
    print("Признаки с нулевой значимостью:")
    print(no_good)

    return no_good
```

#### Encoding & Standard

```python
# Кодирование категориальных признаков трейн и тест (тест на обученном препроцессоре)
def encode_category_feature(X_train, X_test, category_feature):
    encoder = OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore')

    # Строка со списком категориальных признаков
    category_feature_list = category_feature

    # Применение OneHotEncoder
    encoded_train = encoder.fit_transform(X_train[category_feature_list])

    # Сразу кодируем test
    encoded_test = encoder.transform(X_test[category_feature_list])

    # Преобразование результата обратно в DataFrame
    encoded_train_df = pd.DataFrame(encoded_train, columns=encoder.get_feature_names_out(category_feature_list))
    encoded_test_df = pd.DataFrame(encoded_test, columns=encoder.get_feature_names_out(category_feature_list))

    # Сброс индексов
    X_train.reset_index(drop=True, inplace=True)
    X_test.reset_index(drop=True, inplace=True)

    encoded_train_df.reset_index(drop=True, inplace=True)
    encoded_test_df.reset_index(drop=True, inplace=True)

    # Объединение с сбросом индексов
    X_train_encoded = pd.concat([X_train, encoded_train_df], axis=1)
    X_test_encoded = pd.concat([X_test, encoded_test_df], axis=1)

    # Удаление оригинального столбца
    X_train_encoded = X_train_encoded.drop(columns=category_feature)
    X_test_encoded = X_test_encoded.drop(columns=category_feature)


    return X_train_encoded, X_test_encoded




def standard(X_train, X_test, numeric):
    # Применяем StandardScaler к X_train, numeric
    st_scaler = StandardScaler()
    X_train[numeric] = st_scaler.fit_transform(X_train[numeric])
    # Применяем StandardScaler к X_test, numeric
    X_test[numeric] = st_scaler.transform(X_test[numeric])

    return X_train, X_test
```

#### Modeling

```python
def modeling(model_name, X_train_encoded, y_train, X_test_encoded, y_test):
    # Инициализация модели
    model = model_name

    # Обучение модели
    model.fit(X_train_encoded, y_train)

    # Получим предсказания и вероятности
    y_train_pred = model.predict(X_train_encoded)
    y_test_pred = model.predict(X_test_encoded)

    y_train_proba = model.predict_proba(X_train_encoded)
    y_test_proba = model.predict_proba(X_test_encoded)

    # Расчёт метрик ROC AUC
    roc_auc_train = roc_auc_score(y_train, y_train_proba[:, 1])  # Метрика для обучающей выборки
    roc_auc_test = roc_auc_score(y_test, y_test_proba[:, 1])      # Метрика для тестовой выборки

    # Расчёт кросс-валидации для ROC AUC
    roc_auc_cv = cross_val_score(model, X_train_encoded, y_train, cv=5, scoring='roc_auc').mean()

    # Расчёт других метрик
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_test_pred)

    recall_train = recall_score(y_train, y_train_pred)
    recall_test = recall_score(y_test, y_test_pred)

    f1_train = f1_score(y_train, y_train_pred)
    f1_test = f1_score(y_test, y_test_pred)

    # Расчёт кросс-валидации для F1
    f1_cv = cross_val_score(model, X_train_encoded, y_train, cv=5, scoring='f1').mean()

    # Вычисление матрицы ошибок
    cm = confusion_matrix(y_test, y_test_pred)

    # Значения из матрицы ошибок
    tn, fp, fn, tp = cm.ravel()

    # Создадим сводную таблицу для результатов
    df_result = pd.DataFrame({
        'model_name': [model_name.__class__.__name__],  # Название модели
        'roc_auc_train': [roc_auc_train],                # ROC AUC для обучающей выборки
        'roc_auc_test': [roc_auc_test],                  # ROC AUC для тестовой выборки
        'roc_auc_cv': [roc_auc_cv],                      # ROC AUC кросс-валидации
        'accuracy_train': [accuracy_train],              # Accuracy для обучающей выборки
        'accuracy_test': [accuracy_test],                # Accuracy для тестовой выборки
        'recall_train': [recall_train],                  # Recall для обучающей выборки
        'recall_test': [recall_test],                    # Recall для тестовой выборки
        'f1_train': [f1_train],                          # F1 для обучающей выборки
        'f1_test': [f1_test],                            # F1 для тестовой выборки
        'f1_cv': [f1_cv],                                # F1 кросс-валидации
        'tn': [tn],                                      # True Negatives
        'fp': [fp],                                      # False Positives
        'fn': [fn],                                      # False Negatives
        'tp': [tp]                                       # True Positives
    })

    return df_result




# Моделирование с выбором лучшей модели и подбором гиперпараметров
def objective(trial):
    # Выбор классификатора
    classifier_name = trial.suggest_categorical('classifier', ['RandomForest', 'LGBM', 'CatBoost', 'XGBoost'])
    if classifier_name == 'RandomForest':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        max_depth = trial.suggest_int('max_depth', 1, 32)
        clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)

    elif classifier_name == 'LGBM':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        clf = LGBMClassifier(n_estimators=n_estimators, learning_rate=learning_rate)

    elif classifier_name == 'CatBoost':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        depth = trial.suggest_int('depth', 1, 16)
        clf = CatBoostClassifier(n_estimators=n_estimators, learning_rate=learning_rate, depth=depth, verbose=0)

    else:  # XGBoost
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        max_depth = trial.suggest_int('max_depth', 1, 32)
        clf = XGBClassifier(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth, use_label_encoder=False, eval_metric='logloss')

    # Кросс-валидация
    scores = cross_val_score(clf, X_train_encoded, y_train, cv=5, scoring='roc_auc')  # 5-кратная кросс-валидация

    return scores.mean()  # Возвращаем средний roc_auc




# Моделирование с выбором лучшей модели и подбором гиперпараметров
def objective2(X_train_encoded, y_train, X_test_encoded, y_test, trial):
    # Вычисление class_weights для балансировки классов
    class_weights = 'balanced' if len(np.unique(y_train)) == 2 else None

    # Выбор классификатора
    classifier_name = trial.suggest_categorical('classifier', ['RandomForest', 'LGBM', 'CatBoost', 'XGBoost'])
    if classifier_name == 'RandomForest':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        max_depth = trial.suggest_int('max_depth', 1, 32)
        clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, class_weight=class_weights)

    elif classifier_name == 'LGBM':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        clf = LGBMClassifier(n_estimators=n_estimators, learning_rate=learning_rate, class_weight=class_weights)

    elif classifier_name == 'CatBoost':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        depth = trial.suggest_int('depth', 1, 16)
        clf = CatBoostClassifier(n_estimators=n_estimators, learning_rate=learning_rate, depth=depth,
                                 class_weights=class_weights, verbose=0)

    else:  # XGBoost
        n_estimators = trial.suggest_int('n_estimators', 50, 200)
        learning_rate = trial.suggest_loguniform('learning_rate', 1e-3, 1)
        max_depth = trial.suggest_int('max_depth', 1, 32)
        clf = XGBClassifier(n_estimators=n_estimators, learning_rate=learning_rate, max_depth=max_depth,
                            use_label_encoder=False, eval_metric='logloss', scale_pos_weight=(np.sum(y_train == 0) / np.sum(y_train == 1)) if (np.sum(y_train == 1) > 0) else 1)

    # Кросс-валидация
    scores = cross_val_score(clf, X_train_encoded, y_train, cv=5, scoring='roc_auc')
    roc_auc_cv = scores.mean()

    # Обучение модели для получения обучающих метрик
    clf.fit(X_train_encoded, y_train)

    # Предсказание на тестовой выборке для метрик
    y_train_pred = clf.predict(X_train_encoded)
    y_test_pred = clf.predict(X_test_encoded)
    y_probs = clf.predict_proba(X_test_encoded)[:, 1]

    # Метрики
    roc_auc_train = roc_auc_score(y_train, clf.predict_proba(X_train_encoded)[:, 1])
    roc_auc_test = roc_auc_score(y_test, y_probs)
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_test_pred)
    recall_train = recall_score(y_train, y_train_pred)
    recall_test = recall_score(y_test, y_test_pred)
    f1_train = f1_score(y_train, y_train_pred)
    f1_test = f1_score(y_test, y_test_pred)

    tn, fp, fn, tp = confusion_matrix(y_test, y_test_pred).ravel()

    # Создадим сводную таблицу для результатов
    df_result = pd.DataFrame({
        'model_name': [clf.__class__.__name__],  # Название модели
        'roc_auc_train': [roc_auc_train],  # ROC AUC для обучающей выборки
        'roc_auc_test': [roc_auc_test],  # ROC AUC для тестовой выборки
        'roc_auc_cv': [roc_auc_cv],  # ROC AUC кросс-валидации
        'accuracy_train': [accuracy_train],  # Accuracy для обучающей выборки
        'accuracy_test': [accuracy_test],  # Accuracy для тестовой выборки
        'recall_train': [recall_train],  # Recall для обучающей выборки
        'recall_test': [recall_test],  # Recall для тестовой выборки
        'f1_train': [f1_train],  # F1 для обучающей выборки
        'f1_test': [f1_test],  # F1 для тестовой выборки
        'tn': [tn],  # True Negatives
        'fp': [fp],  # False Positives
        'fn': [fn],  # False Negatives
        'tp': [tp]  # True Positives
    })

    # Визуализация ROC-кривой
    fpr, tpr, thresholds = roc_curve(y_test, y_probs)
    roc_auc = auc(fpr, tpr)

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.show()

    # Сохраним параметры модели
    params = {
        'classifier': classifier_name,
        'n_estimators': n_estimators,
        'learning_rate': learning_rate if classifier_name in ['LGBM', 'CatBoost', 'XGBoost'] else None,
        'max_depth': max_depth if classifier_name in ['RandomForest', 'XGBoost'] else None,
        'depth': depth if classifier_name == 'CatBoost' else None,
    }

    # Сохранение атрибутов в trial для последующего извлечения
    trial.set_user_attr("best_model_params", params)
    print(params)
    print(df_result)
    # Возвращаем средний roc_auc кросс-валидации для оптимизации
    return roc_auc_cv
```

#### Подготовка данных.

```python
# удалим id и дубликаты, target==0
df = duplicates(df)

# Предположем пропуск, отсутствием значения
df.fillna(0, inplace=True)

# определение признаков с низкой корреляцией [-0.01: 0.01]
min_corr = col_min_corr(df)
print(len(min_corr))

# Признаки с нулевой значимостью
no_good = importantes_0(df)
print(len(no_good))

# Преобразуем списки в множества, чтобы убрать дубликаты
cols_to_remove = set(min_corr) | set(no_good)

# оставим признаки, необходимые для feature engineering
col_from_features = ['pre_since_opened',
                     'pre_since_confirmed',
                     'pre_till_fclose',
                     'pre_till_pclose',
                     'enc_paym_0',
                     'enc_paym_1',
                     'enc_paym_2',
                     'pre_loans5',
                     'pre_loans530',
                     'pre_loans3060',
                     'pre_loans6090',
                     'pre_loans90',
                     'pre_loans_credit_limit',
                     'pre_maxover2limit',
                     'max_overdue_ratio',
                     'rn'
                     ]
cols_to_remove = [x for x in cols_to_remove if x not in col_from_features]

# Удаляем признаки с нулевой важностью, оставив признаки для features
df.drop(columns=cols_to_remove, inplace=True)
print(df.shape)
print(df['target'].value_counts())

# Повторное удаление дубликатов, target==0
df = duplicates(df)

# Features

# Возраст кредита от открытия до подтверждения 0
df['credit_age'] = df['pre_since_opened'] - df['pre_since_confirmed']

# Остаток времени до закрытия 0
df['remaining_days'] = df['pre_till_fclose'] - df['pre_till_pclose']

# Суммарное состояние платежей за последние 3 месяца
df['payment_status_last'] = df[['enc_paym_0', 'enc_paym_1', 'enc_paym_2']].sum(axis=1)

# Суммарное количество просрочек 0
df['total_overdue_loans'] = df[['pre_loans5', 'pre_loans530', 'pre_loans3060', 'pre_loans6090', 'pre_loans90']].sum(axis=1)

# Проверяем, чтобы избежать деления на ноль
df['max_overdue_ratio'] = np.where(df['pre_loans_credit_limit'] != 0,
                                    df['pre_maxover2limit'] / df['pre_loans_credit_limit'],
                                    0)

# Приведение новых признаков к бинарному виду Признак: высокий уровень просрочки
df['high_overdue_ratio'] = (df['max_overdue_ratio'] > 0.1).astype(int)

# Количество кредитов превышающее среднее значение
# Вычисляем среднее значение 'rn'
mean_rn = df['rn'].mean()

# Создаем новый признак: 'is_recent_credit'
df['is_recent_credit'] = (df['rn'] > mean_rn).astype(int)

# Удаление признаков после feature engineering
df.drop(col_from_features, axis=1, inplace=True)

# Произведем повторную оценку значимости признако
no_good = importantes_0(df)

# Удалим шумы
df.drop(no_good, axis=1, inplace=True)

# Окончательное удаление дубликатов, target==0
df = duplicates(df)
print(df.shape)
```

Кодирование признаков

```python
# Условное разделение. Создадим переменные, хранящие категориальные и числовые признаки
# по принципу: до 7 значений - категориальный

category = []
numeric = []
df_types = df.drop('target', axis=1)
for i in df_types.columns:
    if len(df[i].value_counts()) <= 7:
        category.append(i)
    else:
        numeric.append(i)

print('category:', len(category), 'numeric:', len(numeric))

# Разделяем на X и y
X = df.drop('target', axis=1)
y = df['target']

# Разделяем на обучающую и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Кодируем сразу train и test
X_train_encoded, X_test_encoded = encode_category_feature(X_train, X_test, category)

# Стандартизируем числовые категориальные признаки с большим количеством уникальных значений
X_train_encoded, X_test_encoded = standard(X_train_encoded, X_test_encoded, numeric)
X_train_encoded.head()
```

## Выводы из проведенных исследований и экспериментов:

#### 1. Получение данных будет производится в отдельном модуле.

На этапе получения данных, обработка пропусков производится не будет. Поскольку остается последующая вероятность обработки выбросов. Заполнение пропусков до обработки выбросов и аномалий может исказить данные.

#### 2. После получения данных, произведем оценку важности признаков.

Для реализации используем SHAP и XGBoost. Удаление признаков с нулевой важностью позволит существенно сократить вычислительное время.

Дополнительно это позволит сократить количество шумов в данных, что улучшит точность модели.

#### 3. Data cleaning

Удаление признаков с нулевой важностью, удаление id.

Предусмотрено заполнение пропусков - 0. Предположим, что пропуски обусловлены отсутствием значения.

Выбросы не обрабатываются, поскольку вероятнее всего, объекты малого класса, определяются как выбросы.

При обработке дубликатов, оставлены дубликаты меньшего класса, что позволит немного сократить дисбаланс.

#### 4. Feature engineering

Созданы новые признаки:

Высокий уровень просрочки.

Суммарное количество просрочек.

Суммарное состояние платежей за последние 3 месяца.

Возраст кредита от открытия до подтверждения.

Остаток времени до закрытия кредита.

#### 5. Кодирование и стандартизация.

Произведено условное разделение признаков на категориальные и непрерывные.

По принципу количества уникальных значений. Установим порог = 9. Менее 10 уникальных значений признака - category, более 10 - numeric.

Данное разделение позволит существенно сократить вычислительное время при кодировании категориальных признаков.

Кодирование производилось отдельно для train и test. На train обучался OHE, test кодировался на обученом OHE.

Для остальных признаков использовался StandardScaler.

В экспериментах производился подбор весов классов и балансировка классов.

Кросс-валидация на данном этапе не производилась, ввиду экономии процесорного времени
